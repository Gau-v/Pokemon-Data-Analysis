Aggregate statistics(Groupby)

```python
df = pd.read_csv('D:\Sample Folder\pokemon_data.csv')
# Select only the numeric columns for aggregation
numeric_columns = df.select_dtypes(include=['number']).columns

# Group by 'Type 1', calculate the mean for numeric columns, and sort by 'HP' in descending order
df_grouped = df.groupby(['Type 1'])[numeric_columns].mean().sort_values('HP', ascending=False)

print(df_grouped.head(5))
```
